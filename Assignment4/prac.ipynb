```python
import pandas as pd
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
X = df.drop(columns=['Outcome'])
Y = df['Outcome']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test)
```

```python
cmf = confusion_matrix(Y_test, Y_pred)
```

```python
ConfusionMatrixDisplay(cmf).plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```
